```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://byjus.com/cbse/essay-on-independence-day/'
req = requests.get(url)
htmlPage = req.content
# print(htmlPage) will gives you unstructed html page
```

```python
#structered page
soup = BeautifulSoup(htmlPage, 'html.parser')
soup
```

```python
print(soup.prettify()) #to pretify more with indentation
```

```python
print(soup.title) #title of page with tag
print(soup.title.string) #only title of page
```

```python
#different types of objects 
print(type(soup))
print(type(soup.title))
print(type(soup.title.string))
```

```python
#all paragraphs
print(soup.find_all('p'))
```

```python
#all anchor tags
print(soup.find_all('a'))
```

```python
print(soup.find('a')) #first anchor tag
print(soup.find('p')) #first para
```

```python
print(soup.find('a')['class']) #class of the first anchor tag
```

```python
print(soup.find_all('a', class_='close')) #list of anchor tags having class as 'close'
```

```python
print(soup.find('p').get_text()) #get only content of para
```

```python
#get links from all anchor tags
anchors = soup.find_all('a')
for link in anchors:
    print(link.get('href'))
```

```python
id_content = soup.find(id ='post-1420741')
```

```python
print(id_content)
```

```python
print(id_content.contents) #gives list
```

```python
for element in id_content.children:
    print(element)
```

```python
# .contents = is list
# .children = is generator
```

```python
for string in id_content.strings:
    print(string)
```

```python
for string in id_content.stripped_strings:
    print(string)
    
#remove unnessary spaces
```

```python
print(id_content.parent)
```

```python
for parent in id_content.parents:
    print(parent.name)
```

```python
print(id_content.previous_sibling)
```

```python
print(id_content.next_sibling.next_sibling)
```

```python
print(soup.select('.clearfix')) #list of elements having class as 'clearfix'
```

```python
print(soup.select('#city')) #list of elements having id as 'city'
```

```python
print(soup.find_all('div'))
```

```python
url = 'https://www.imdb.com/chart/top/'
```

```python
pageContent = requests.get(url).content
movieSoup = BeautifulSoup(pageContent, 'html.parser')
```

```python
movies = movieSoup.find_all('td', class_='titleColumn')
print(len(movies))
```

```python
movies[0]
```

```python
#get names and year
for movie in movies:
    print(movie.a.string,movie.span.string )
```
